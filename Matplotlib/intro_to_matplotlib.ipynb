## Matplotlib example workflow 

```python
# 0. import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# 1. get some data ready
x = [1,2,3,4,5]
y = [11,12,13,14,15]

# 2. setup object oriented plot
fig, ax = plt.subplots(figsize=(5,5)) #width and height of the subplot

# 3. plot the data
ax.plot(x,y)

#4. customize the plot
ax.set(title="Simple Plot",
      xlabel = "x-axis",
      ylabel = "y-axis")

#5. Save the plot
fig.savefig("images/sample-plot.png")
```

```python
## Making a horizontal Bar plot from a dictionary
nut_butter_prices = {"Almond":10,"Peanut":8,"Cashew":12}
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));

ax.set(title ="Nut Butter Prices",
      xlabel = "Price ($)");
```

### Making subplots (multiple plots in the same canvas)
### TWO OPTIONS FOR MAKING THEM

```python
# Method # 1
fig , ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
#its pretty self-explanatory

#adding data to subplots
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Method # 2
fig , ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
#its pretty self-explanatory

#adding data to indices of instance of subplots
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

## Which method to use for plots (pyplot vs matplotlib OOP method)
* When plotting something quickly and roughly then use pyplot
* When plotting something more advanced and accurate, use OOP method instead

```python
heart_disease.head()
```

```python
len(heart_disease)
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
len(over_50)
```

```python
#pyplot method
over_50.plot(x="age",y="chol",c="target",kind="scatter");
```

```python
#matplotlib OO method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(x="age",y="chol",c="target",kind="scatter",ax=ax);

#increasing the limit of x-axis
ax.set_xlim([45,100])
```

```python
#matplotlib OO method mixed with pyplot
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(x="age",y="chol",c="target",kind="scatter",ax=ax);

#increasing the limit of x-axis
#ax.set_xlim([45,100])
```

```python
# pur OO method
fig, ax = plt.subplots(figsize=(10,6))

#plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);
#customizing the plot
ax.set(title="Heart Disease and Cholestrol Levels",
       xlabel = "Age",
       ylabel="Cholestrol Levels");

#add a legend
ax.legend(*scatter.legend_elements(),title="Target")
# it takes the legend element from scatter that is "target" and looks up the values in the data set and uses them for the legend

#add a horizontal line
ax.axhline(over_50["chol"].mean(),linestyle="--");
```

```python
#creating subplots
# pur OO method

#they will have the same x-axis values
fig, (ax1,ax2) = plt.subplots(figsize=(10,10),nrows=2,sharex=True)

#plot the data
scatter_chol = ax1.scatter(
                     x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]
                            );
scatter_thalach = ax2.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"]);
#customizing the plot
ax1.set(title="Heart Disease and Cholestrol Levels",
       ylabel="Cholestrol Levels");
ax2.set(title="Heart Disease and Max Heart Rate",
       xlabel = "Age",
       ylabel="Max Heart Rate");

#add a legend
ax1.legend(*scatter_chol.legend_elements(),title="Target")
ax2.legend(*scatter_thalach.legend_elements(),title="Target")
# it takes the legend element from scatter that is "target" and looks up the values in the data set and uses them for the legend

#add a horizontal line
ax1.axhline(over_50["chol"].mean(),linestyle="--");
ax2.axhline(over_50["thalach"].mean(),linestyle="--");

#giving a title to the entire figure
fig.suptitle("Heart Disease Analysis",fontsize = 18,fontweight="bold",verticalalignment="center");
```

## Changing style within a style

```python
#change the style
plt.style.use('seaborn-whitegrid')
# pur OO method
fig, ax = plt.subplots(figsize=(10,6))

#plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap='winter'); # cmap is color scheme, you can change it to change the colors in the graph
#customizing the plot
ax.set(title="Heart Disease and Cholestrol Levels",
       xlabel = "Age",
       ylabel="Cholestrol Levels");

#add a legend
ax.legend(*scatter.legend_elements(),title="Target")
# it takes the legend element from scatter that is "target" and looks up the values in the data set and uses them for the legend

#add a horizontal line
ax.axhline(over_50["chol"].mean(),linestyle="--");
```

```python
#customizing x and y limitations

#change the style
plt.style.use('seaborn-whitegrid')

#they will have the same x-axis values
fig, (ax1,ax2) = plt.subplots(figsize=(10,10),nrows=2,sharex=True)

#plot the data
scatter_chol = ax1.scatter(
                     x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap='winter'
                            );
scatter_thalach = ax2.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],cmap='winter');
#customizing the plot
ax1.set(title="Heart Disease and Cholestrol Levels",
       ylabel="Cholestrol Levels");
#changing x and y limits of first graph
ax1.set_xlim([50,80]);
ax1.set_ylim([150,400]);

ax2.set(title="Heart Disease and Max Heart Rate",
       xlabel = "Age",
       ylabel="Max Heart Rate");
#changing x and y limits of second graph
ax2.set_xlim([50,80]);
ax2.set_ylim([60,200]);

#add a legend
ax1.legend(*scatter_chol.legend_elements(),title="Target")
ax2.legend(*scatter_thalach.legend_elements(),title="Target")
# it takes the legend element from scatter that is "target" and looks up the values in the data set and uses them for the legend

#add a horizontal line
ax1.axhline(over_50["chol"].mean(),linestyle="--");
ax2.axhline(over_50["thalach"].mean(),linestyle="--");

#giving a title to the entire figure
fig.suptitle("Heart Disease Analysis",fontsize = 18,fontweight="bold",verticalalignment="center");
```
